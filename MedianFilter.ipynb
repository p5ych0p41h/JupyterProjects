```python
from PIL import Image
import matplotlib.pyplot as plt
```

### Открываем изображение

```python
image = Image.open("image1.png")
```

### Обработка изображения

```python
#Сохраняем ширину и высоту изображения
width, height = image.size

#Создаем пустое изображения с тем же размером
filteredImage = Image.new("RGB", (width, height))

#Проход по каждому пикселю
for x in range(width):
    for y in range(height):
        #Создание матрицы обработки 3х3
        pixels = []
        #Проход по пикселям матрицы обработки
        for i in range(-1, 2):
            for j in range(-1, 2):
                #Проверка выхода за границы изображение
                if 0 <= x+i < width and 0 <= y+j < height:
                    pixel = image.getpixel((x+i, y+j))
                    pixels.append(pixel)
        #Сортировка матрицы обработки
        pixels.sort()
        #Поиск медианы
        medianPixel = pixels[len(pixels)//2]
        #Установка медианы в пиксель
        filteredImage.putpixel((x, y), medianPixel)
```

### Вывод изображения

```python
plt.figure()
f, a = plt.subplots(1, 2)
a[0].imshow(image)
a[1].imshow(filteredImage)
```
